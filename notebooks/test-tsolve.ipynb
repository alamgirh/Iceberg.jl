```julia
using Iceberg
using PyPlot
```

```julia
include("Iceberg.jl")
ms, mpar, ppar = Iceberg.initialize1d(100)
mpar.dt = 1e-5

axls = axes()
x = linspace(0.0, 1.0, mpar.nx[1])
axls[:plot](x, ms.phi, "--k")
axls[:plot](x, zeros(length(x)), ":k")

axT = twinx()

axT[:plot](x, ms.temp, "k")
ylim(-1.5, 1.5)

nt = 1000

for i=1:nt
    t = Iceberg.tsolve(ms, mpar, ppar)
    vel = Iceberg.front_velocity(ms, mpar, ppar)
    ms.phi += mpar.dt * vel
    #Iceberg.lsupdate!(ms, mpar, ppar)
    if (i%50)==0
        axls[:plot](x, ms.phi, "--r", alpha=0.2i/nt)
        axT[:plot](x, t, "-b", alpha=0.5i/nt)
    end
end
```
